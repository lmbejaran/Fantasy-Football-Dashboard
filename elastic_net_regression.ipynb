```python
#import dependencies
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import scipy as sp
import sklearn as sk
from pandas import read_csv
from matplotlib import pyplot
from numpy import mean
from numpy import std
from numpy import absolute
from numpy import arange
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import ElasticNetCV
```

```python
#load fantasy PL data
data = ('api/data/resources/cut_points_data.csv')
```

```python
#check data pull
df = pd.read_csv(data)
df
```

```python
#convert columns to float type
df['Unnamed: 0'] = df['Unnamed: 0'].astype(float)
df['id'] = df['id'].astype(float)
df['form'] = df['form'].astype(float)
df['total_points'] = df['total_points'].astype(float)
df['points_per_game'] = df['points_per_game'].astype(float)
df['event_points'] = df['event_points'].astype(float)
df['goals_scored'] = df['goals_scored'].astype(float)
df['assists'] = df['assists'].astype(float)
df['clean_sheets'] = df['clean_sheets'].astype(float)
df['bonus'] = df['bonus'].astype(float)
df['bps'] = df['bps'].astype(float)
df['influence'] = df['influence'].astype(float)
df['creativity'] = df['creativity'].astype(float)
df['threat'] = df['threat'].astype(float)
df['ict_index'] = df['ict_index'].astype(float)
```

```python
#define model
model = ElasticNet(alpha = 1.0, l1_ratio=0.5)
```

```python
#load data
data = df.values
X, y = data[:, :-1], data[:, -1]
```

```python
#define model
model = ElasticNet(alpha=1.0, l1_ratio=0.5)
```

```python
#define model evaluation method
cv = RepeatedKFold(n_splits=100, n_repeats=30, random_state=1)
```

```python
#evaluate model
scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
```

```python
#force scores to be positive
scores = absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```
